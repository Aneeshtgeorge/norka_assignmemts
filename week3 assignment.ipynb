##  Meaningful visualizations using the dataset

```python
sns.countplot(df1['Classification'])
plt.grid()
```

##### From above graph count equal number of iris-setosa, iris-versicolor, iris-virginica are present(50 each)

```python
sns.pairplot(df1,hue='Classification')
```

#### The above pairplot gives an idea about the overall relationship between "Classification" and other columns.

```python
plt.figure(figsize=(8,8))

plt.subplot(2,2,1)
sns.boxplot(x='Classification',y='PL',data=df1)

plt.subplot(2,2,2)
sns.boxplot(x='Classification',y='PW',data=df1)

plt.subplot(2,2,3)
sns.boxplot(x='Classification',y='SL',data=df1)

plt.subplot(2,2,4)
sns.boxplot(x='Classification',y='SW',data=df1)

plt.tight_layout()
```

#### Above graph shows more statistical explonatory analysis using boxplot
